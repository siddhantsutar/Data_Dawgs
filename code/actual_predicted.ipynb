```python
import json
import numpy as np
import matplotlib.pyplot as plt
import pylab
from sklearn.metrics import mean_squared_error
%matplotlib inline  
```

```python
actual = open("actual.json", "r")
predicted = open("predicted.json", "r")
```

```python
actual_dict = json.load(actual)
predicted_dict = json.load(predicted)
```

```python
x = []
y = []
```

```python
for key, value in predicted_dict.iteritems():
    x.append(value)
    y.append(actual_dict[key])
    if (abs(x[-1] - y[-1]) > 3):
        print key
```

```python
plt.figure(figsize=(15,10))
plt.scatter(x, y, color = 'red')
plt.plot(x, np.poly1d(np.polyfit(x, y, 1))(x)) #line of best-fit
pylab.title('Predicted vs. Actual ratings')
pylab.xlabel('Predicted')
pylab.ylabel('Actual')
plt.savefig('regression_analysis.png')
```

```python
mean_squared_error(x, y)
```
